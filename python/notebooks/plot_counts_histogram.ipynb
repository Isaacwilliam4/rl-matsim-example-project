```python
df21 = df[df['Link Id'] == 21]
df21.head()
```

```python
unique_hours = pd.unique(df21['Hour'])
bar_width = 0.4  # Width of each bar

# X positions for each set of bars
r1 = np.arange(len(unique_hours))
r2 = [x + bar_width for x in r1]

plt.figure(figsize=(12, 6))

# Plot Count volumes histogram
plt.bar(r1, df21['Count volumes'], color='blue', width=bar_width, label='Count volumes')

# Plot MATSIM volumes histogram
plt.bar(r2, df21['MATSIM volumes'], color='orange', width=bar_width, label='MATSIM volumes')

# Labels, title, and legend
plt.xlabel('Hour')
plt.ylabel('Volumes')
plt.title('Comparison of Count Volumes and MATSIM Volumes')
plt.xticks([r + bar_width / 2 for r in range(len(unique_hours))], unique_hours)
plt.legend()

plt.show()
```
